### Import Libraries

```python
import numpy as np
```

### Creation of Dataset using Numpy Array

```python
stuexam = np.array([
    [85, 90, np.nan],
    [72, np.nan, 68],
    [91, 88, 95],
    [np.nan, 77, 82],
    [60, 65, 70],
    [78, 82, np.nan],
    [95, 91, 89],
    [55, np.nan, 60]
])
print(stuexam)
```

### Figuring out the Rows and Columns

```python
print(stuexam.shape)
```

### Count the Missing Values

```python
print(np.isnan(stuexam).sum())
```

### Search of Null value

```python
for i in range(0,8):
     for j in range(0,3):
          if(np.isnan(stuexam[i][j])):
               print("NaN value found at index: ", i, j)
```

### Mean Across Column

```python
stu_avg=np.nanmean(stuexam, axis=1)
print(stu_avg)
```

### Replace Nan value with Mean

```python
data=stuexam.copy() # Create a copy of the original array to avoid modifying it directly
print(data)
for i in range(0,8):
     for j in range(0,3):
          if(np.isnan(data[i][j])):
               data[i][j]=stu_avg[i]
print(data)
```

### Mean across down Row

```python
sub_avg=np.nanmean(stuexam, axis=0) # Calculation of subject-wise average, ignoring NaN values
print(sub_avg)
```

### Who will passed?

```python
passed=stu_avg[stu_avg>=70]
print(len(passed),"Students passed the exam.")
```

### Summary Report

```python
print("Overall Mean:",np.nanmean(stuexam))
print("Highest Score:",np.nanmax(stuexam))
print("Lowest Score:",np.nanmin(stuexam))
print("Student passed :",len(passed),"/8")
np.savetxt('results.csv', data, delimiter=',')
```
